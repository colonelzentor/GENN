```python
from genn.model import GENN
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

# Tutorial 1: One-Dimensional Sinusoid

In a first step, let's define the test function we will be working with, taken to be a simple sinusoidal function:

```python
f = lambda x: x * np.sin(x)
df_dx = lambda x: np.sin(x) + x * np.cos(x)
```

Next, let's define the domain over which we will collect synthetic training data:

```python
lb = -np.pi
ub = np.pi
```

We will now generate some synthetic data that will be used to train our GENN model later on:

```python
m = 4    # number of training examples
n_x = 1  # number of inputs
n_y = 1  # number of outputs
X_train = np.linspace(lb, ub, m).reshape((n_x, m))
Y_train = f(X_train).reshape((n_y, m))
J_train = df_dx(X_train).reshape((n_y, n_x, m))
```

We also generate some synthetic data that will be used to test the accuracy of the trained model:

```python
m = 30  # number of test examples
X_test = lb + np.random.rand(m, 1).reshape((n_x, m)) * (ub - lb)
Y_test = f(X_test).reshape((n_y, m))
J_test = df_dx(X_test).reshape((n_y, n_x, m))
```

We now initalize the model, where $deep \ge 1$ and $wide \ge 1$ are the number of layers in the neural network and the number of nodes per layer, respectively. Feel free to try different network architectures. The parameters $n_x \ge 1$ and $n_y \ge 1$ are the number of inputs and outputs, respectively. 

```python
model = GENN.initialize(n_x=X_train.shape[0], n_y=Y_train.shape[0], deep=2, wide=12)
```

Now that the model is initialized, we need to train it. To do so, we need to provide training data and specify hyper-parameters that control the behavior of the training algorithm:

```python
model.train(X=X_train,
            Y=Y_train,
            J=J_train,
            alpha=0.05,  # optimizer learning rate (control line search step size)
            lambd=0.1,   # regularization coefficient (to avoid over-fitting)
            gamma=1.0,   # gradient-enhancement coefficient (set gamma = 0 to turn off gradient enhancement)
            beta1=0.9,   # tuning parameter for ADAM optimization (recommended value is 0.9)
            beta2=0.99,  # tuning parameter for ADAM optimization (recommended value is 0.99)
            mini_batch_size=None,  # if not None, the data will be broken up into batches of specified size
            num_iterations=100,    # number of optimizer iterations per mini batch
            num_epochs=10)         # number of passes through the data  
```

Now that the model is trained, let's check the results. To do so, we first verify the convergence history: 

```python
model.plot_training_history()
```

Upon satisfaction, we now now check the goodness of fit using the test data we generated earlier: 

```python
model.goodness_of_fit(X_test, Y_test)
```

We can also check the goodness of fit of the derivative:

```python
model.goodness_of_fit(X_test, Y_test, J_test)
```

Finally, since this is a 1D problem, we can plot the predicted and true responses for comparison. In a first step, let's generate new data points for plotting:

```python
X = np.linspace(lb, ub, 100).reshape((n_x, 100))
```

Second, let's evaluate the true and predicted model at these points:

```python
Y_true = f(X)
Y_pred = model.evaluate(X)
```

Finally, we can now plot the true and predicted models for comparison: 

```python
fig, ax = plt.subplots()
ax.plot(X.T, Y_pred.T, 'b-')
ax.plot(X.T, Y_true.T, 'k--')
ax.plot(X_test.T, Y_test.T, 'ro')
ax.plot(X_train.T, Y_train.T, 'k+', mew=3, ms=10)
ax.set(xlabel='x', ylabel='y', title='GENN')
ax.legend(['Predicted', 'True', 'Test', 'Train'])
plt.show()
```

Notice that with only four training points, we obtained a pretty descent prediction. Now, for comparison, let's turn off gradient enhancement (gamma = 0) and repeat these steps: 

```python
model.train(X=X_train,
            Y=Y_train,
            J=J_train,
            alpha=0.05,  # optimizer learning rate (control line search step size)
            lambd=0.1,   # regularization coefficient (to avoid over-fitting)
            gamma=0.0,   # gradient-enhancement coefficient (set gamma = 0 to turn off gradient enhancement)
            beta1=0.9,   # tuning parameter for ADAM optimization (recommended value is 0.9)
            beta2=0.99,  # tuning parameter for ADAM optimization (recommended value is 0.99)
            mini_batch_size=None,  # if not None, the data will be broken up into batches of specified size
            num_iterations=100,    # number of optimizer iterations per mini batch
            num_epochs=10)         # number of passes through the data  
Y_true = f(X)
Y_pred = model.evaluate(X)
fig, ax = plt.subplots()
ax.plot(X.T, Y_pred.T, 'b-')
ax.plot(X.T, Y_true.T, 'k--')
ax.plot(X_test.T, Y_test.T, 'ro')
ax.plot(X_train.T, Y_train.T, 'k+', mew=3, ms=10)
ax.set(xlabel='x', ylabel='y', title='NN')
ax.legend(['Predicted', 'True', 'Test', 'Train'])
plt.show()
```

You can see that even though we perfectly fit the training data, the prediction is actually terrible because, without gradient-enhancement, tradional neural nets need more data points to learn the topology of the response they are trying to fit.
